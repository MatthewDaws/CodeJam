```python
def is_palindrome(num):
    s = str(num)
    return s == "".join(reversed(s))

is_palindrome(121)
```

```python
for n in range(1,1000000):
    if is_palindrome(n) and is_palindrome(n*n):
        print(n, n*n)
```

```python
for length in range(1,15):
    sublen = (length + (length %2)) // 2
    fair_and_squares = []
    for n in range(1,10**sublen):
        s = str(n)
        s = "0"*(sublen-len(s)) + s
        if length % 2 == 1:
            s = "".join(reversed(s[1:])) + s
        else:
            s = "".join(reversed(s)) + s
        if s[0] != "0":
            num = int(s)
            if is_palindrome(num*num):
                fair_and_squares.append(num)
                
    fair_and_squares.sort()
    for n in fair_and_squares:
        print(n,n*n)
```

```python
def find_fas(length):
    sublen = (length + (length %2)) // 2
    fair_and_squares = []
    for n in range(1,10**sublen):
        s = str(n)
        s = "0"*(sublen-len(s)) + s
        if length % 2 == 1:
            s = "".join(reversed(s[1:])) + s
        else:
            s = "".join(reversed(s)) + s
        if s[0] != "0":
            num = int(s)
            if is_palindrome(num*num):
                fair_and_squares.append(num)  
    fair_and_squares.sort()
    return fair_and_squares

find_fas(14)
```
